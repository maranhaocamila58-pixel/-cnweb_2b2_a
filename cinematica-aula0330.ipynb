# Fórmula da velocidade

```
def calculaVelMedia(deltaS, deltaT):
  VelMedia = (deltaS / deltaT)
  return VelMedia
```

## Teste Serviço Nuvem: Velocidade Média

```
print("Vm=", calculaVelMedia(30,15))
```

#Fórmula da aceleração

```
def CalculaAceleracaoMedia(deltaV, deltaT):
  aceleracaoMedia = (deltaV / deltaT)
  return aceleracaoMedia
```

#Teste de Serviço Nuvem:Aceleração Média

```
print("Am =", CalculaAceleracaoMedia(30,15))
```

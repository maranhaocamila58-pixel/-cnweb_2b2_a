```
import math

def equacacao2grau(a, b, c):
  delta = b**2 - 4*a*c
  if delta < 0:
    return "sem raiz para este número"
  x1 = (-b + math.sqrt(delta))/ (2*a)
  x2 = (-b - math.sqrt(delta)) / (2*a)
  return x1, x2
```

```
print(equacacao2grau(1,-3,2))
```
